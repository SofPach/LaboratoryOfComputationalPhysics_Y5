7. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).

Constract a shape (100,) boolean array, the mask
Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
Apply the mask to obtain an array of ordered prime numbers
Check the performances (timeit); how does it scale with N?
Implement the optimization suggested in the sieve of Eratosthenes

(20 mins)

````python
def prime1(N):
    a=np.arange(0,N)
    mask=np.array([True]*N)
    for i in range(2, N ):
        mult=[]
        for j in range(i,N,i):
            mult.append(j)
        mask[mult[1:]]=False
    return a[mask][2:]
````

````python
timeit(prime1(100))
````

````python
prime1(100)
````

````python
def prime2(N):
    b=np.arange(0,N)
    mask=np.array([True]*N)
    for i in np.arange(2,np.sqrt(N),1):
        i=int(i)
        if mask[i]==True:
            lis=[(i*2+i*k) for k in range(N)]
            for j in lis:
                if j<N:
                    mask[j]=False
    return b[mask][2:]
````

````python
prime2(100)
````

````python
timeit(prime2(100))
````

* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg as la
````

````python
#Dataset with 3 features, each with N entries(N=O 1000), N(mu,sigma)
#Generate 3 variables xi: x_1 ~ N(0,1), x_2 ~ (x_1+N(0,3)), x_3~(2x_1+x_2)

n= 1000
x1 = np.random.normal(loc=0.0, scale=1, size=n)
x2 = x1 + np.random.normal(loc=0.0, scale=3, size=n)
x3 = 2*x1 + x2

X = np.array([x1,x2,x3])
X
````

````python
#Find the eigenvectors and eigenvalues of the covariance matrix of the dataset

l, V = la.eig(np.cov(X))
print(l)
print(V)
````

````python
#Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result

U, spectrum, Vt = np.linalg.svd(X)
l_svd = (spectrum**2)/(n-1)
V_svd = U
print(l_svd)
print(V_svd)
````

````python
#What percent of total variability is explained by the principal components? 
#Given how the dataset was constructed, do these make sense?
#Reduce the dimensionality of the system so that at least 99% of the total variability is retained.

percent=(l_svd[0]+l_svd[1])/sum(l_svd)
print('More or less we got a', percent*100, '% of the variability explained by the principal components.')
print('It makes sense because of being a normal distribution')
````

````python
#Redefine the data in the basis yielded by the PCA procedure
spectrum[-1]=0

m, n = U.shape[0], Vt.shape[0]
D = np.zeros((m, n))
# U = mxm, spectrum = mxn, Vt = nxn
for i in range(min(m, n)):
    D[i, i] = spectrum[i] 

#Redefined data using dimensionality reduction by SVD
X_new = np.dot(U, np.dot(D, Vt))
X_new
````

````python
#Plot the data points in the original and the new coordiantes as a set of scatter plots. 
#Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.

fig = plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.scatter(X[0], X[1], alpha=0.2)
plt.title("(0,1) projection",fontsize=14)
plt.ylabel("X dataset original", fontsize=16)

plt.subplot(2,3,2)
plt.scatter(X[0], X[2], alpha=0.2)
plt.title("(0,2) projection",fontsize=14)

plt.subplot(2,3,3)
plt.scatter(X[1], X[2], alpha=0.2)
plt.title("(1,2) projection",fontsize=14)

plt.subplot(2,3,4)
plt.scatter(X_new[0], X_new[1], alpha=0.2)
plt.ylabel("X_new dataset redefined",fontsize=16)

plt.subplot(2,3,5)
plt.scatter(X_new[0], X_new[2], alpha=0.2)

plt.subplot(2,3,6)
plt.scatter(X_new[1], X_new[2], alpha=0.2)

plt.show()
````
